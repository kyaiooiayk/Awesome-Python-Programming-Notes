# Solution #1
<hr style = "border:2px solid black" ></hr>

- [ ] Was not run on leetcode autograder as it required a solution with linked list.

````python
def mergeTwoLists(list1: list[int], list2: list[int]) -> list[int]:
    
    if list1 == []:
        return list2
    if list2 == []:
        return list1
    if list1 == [] and list2 == []:
        return []

    else:
        return sorted(list1 + list2)

print(mergeTwoLists(list1 = [1,2,4], list2 = [1,4,3]), "Expected=: [1,1,2,3,4,4]")
print(mergeTwoLists(list1 = [], list2 = []), "Expected=: []")
print(mergeTwoLists(list1 = [], list2 = [0]), "Expected=: [0]")
````

# Create singly linked lists
<hr style = "border:2px solid black" ></hr>

- [x] Passed Leetcode autograder!

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
````

````python
def print_list(list1):
    temp = True
    list_ = []

    while list1:
        list_.append(list1.val)
        list1 = list1.next
    return list_
````

````python
print(print_list(ListNode(1, ListNode(2, ListNode(4)))))
print(print_list(ListNode(1, ListNode(4, ListNode(3)))))
print(print_list(ListNode(0)))
print(print_list(ListNode())) #?
````

# Solution #2
<hr style = "border:2px solid black" ></hr>

````python
def mergeTwoLists(list1, list2):

    if list1 != None and list2 == None:
        return list1
    elif list1 == None and list2 != None:
        return list2
    elif list1 == None and list2 == None:
        return None
    elif list1.val <= list2.val:
        return ListNode(list1.val, mergeTwoLists(list1.next, list2))
    else:
        return ListNode(list2.val, mergeTwoLists(list1, list2.next))
````

````python
list1 = ListNode(1, ListNode(2, ListNode(4)))
list2 = ListNode(1, ListNode(4, ListNode(3)))
list3 = ListNode(0)
list4 = ListNode(0)
````

````python
print(print_list(mergeTwoLists(list1, list2)), "Expected: [1,1,2,4,4,3]")
print(print_list(mergeTwoLists(list4, list3)), "Expected: [0,0]")
````

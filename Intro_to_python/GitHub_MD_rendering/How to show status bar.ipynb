# Import modules

```python
from tqdm import tqdm
```

# Without status bar

```python
a = []
import numpy as np
b = np.linspace(0,10,int(1.e2))

for dummy in b:
    a.append(dummy)
print(a)
```

# tqdm to show status bar

```python
a = []
import numpy as np
b = np.linspace(0,10,int(1.e7))

for dummy in tqdm(b, desc = "Reading and appending"):
    a.append(dummy)
print(a[:10])
```
